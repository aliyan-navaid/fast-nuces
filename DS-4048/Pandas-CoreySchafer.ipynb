# 6- Create, Update, Delete

## 6.1 Index

```python
schema = schema.reset_index()
schema = schema.set_index("qname")
schema.sort_index(ascending=True, inplace=False).head(5)
```

## 6.2 Columns

```python
# Rename All
schema.columns = ["qid", "question", "type", "sub", "sq_id"]
schema.columns = [ x.upper() for x in schema.columns ]
schema.columns = schema.columns.str.upper() # equivalent
schema.columns = schema.columns.str.replace(" ", "_")

schema.columns
```

```python
# Rename Selected
df = df.rename( columns={'EdLevel':'Education_Level'} )
df.columns
```

```python
# Add Columns
schema["exp_column"] = schema["QUESTION"] + ' ' + schema["SUB"] # Concatenate 2 string columns
schema["split2strings"] = schema["QUESTION"].str.split(' ')

schema.head()
```

```python
# Drop Columns

schema = schema.drop( columns=["exp_column", "split2strings"] )
```

### 6.2.1 Apply

### 6.2.2 Map
Similar to Apply except for DF

```python
# Series - Applies Function on Cells

df["Country"].map(type)
```

```python
# DF - Applies Function on All Cells
df.map(type)
```
