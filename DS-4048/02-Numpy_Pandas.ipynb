```python
max_idx = df5.index.max()
df6_tmp = df6.copy()

df6_tmp.index = range(max_idx + 1, max_idx + 1 + len(df6))


df7 = pd.concat([df5, df6_tmp], sort=False)
display(df5, df6, df7)
```

```python
df7.replace("Hello", np.nan, inplace=True)
df7
```

```python
df7["A"] = pd.to_numeric(df7['A'], errors='coerce')
df8 = df7.fillna(df7.mean(numeric_only=True))
df8
```

# Question 2 - List Operations

```python
first: list = [11.25, 18.0, 20.0, 10.75, 9.5]
second: list= [10.75, 9.50]

full = first+second
print(f"First: {first}")
print(f"Second: {second}")
print(f"First + Second: {full}")

full_sorted = sorted(full, reverse=True) # slightly more efficient but not inplace unlike list.sort()
print(f"Full Sorted: {full_sorted}")

full_sorted.reverse()
print(f"Reversed Full Sorted: {full_sorted}")
```

# Question 4 - NP Array Slicing/Indexing

```python
# dummy data: 7 days x 24 hours
data = np.arange(7*24).reshape(7, 24)

# every other hour of Day 1
a = data[0,::2]
print("Day1 every other hour:", a)

# first 12 hours for all days
b = data[:,:12]
print("First 12 hours all days:\n", b)

# every 6th hour
c = data[:,::6]
print("Every 6th hour all days:\n", c)

# last 3 days
e = data[4:,:]
print("Last 3 days:\n", e)

# reverse Day 3
f = data[2,::-1]
print("Day 3 reversed:", f)

# every 3rd hour, first 4 days
g = data[:4,::3]
print("Every 3rd hour first 4 days:\n", g)
```
