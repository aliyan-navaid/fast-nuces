# Lab Task

```python
df = pd.read_csv("./Datasets/K230686-Bakery.csv", index_col=0)
```

```python
df.head()
```

```python
from tabulate import tabulate


headers = [ "Item", "Initial Cost ($)", "ingr_cost($/hr)", "labour_cost($/hr)", "sale($/hr)" ]
rows = []
for item in df.index:
    initial_cost = df.loc[item, "Initial Cost ($)"]
    ingredient_cost_hourly = df.loc[item, "Ingredient Cost ($/lbs)"]*df.loc[item, "Ingredients per Batch (lbs)"]*df.loc[item, "Batches per Hour"]
    labour_cost_hourly = df.loc[item, "Labor Charge ($/hr)"]
    sale_hourly = df.loc[item, "Batches per Hour"]*df.loc[item, "Items per Batch"]*df.loc[item, "Item Sale Price ($)"]
    rows.append( [item, initial_cost, ingredient_cost_hourly, labour_cost_hourly, sale_hourly] )

print(tabulate(rows, headers=headers, tablefmt="grid"))
print("\n")

print("Initial Cost($)  Profit($/hr)")
for row in rows:
    row[4] -= row[2] + row[3]
    print(f"{row[1]}             {row[4]}")
print("\n")

print("Minuets to recover Initial Investment")
for row in rows:
    if row[4]<=0: 
        print(0)
        continue
    hours = row[4]/row[1]*60
    print(f"{hours:.2f}")
print("\n")

print("Bagel is best item")
```

# 1. Deleting Redundant Columns

```python
df = df.drop( columns=["location", "listed_in(city)"] )
df.head(2)
```

# 2. Renaming Columns

```python
df = df.rename( columns={"rate":"rating", "phone":"phone_number", "rest_type":"restaurant_type"} )
df.head(2)
```

# 3. Dropping Duplicate Rows

```python
df = df.drop_duplicates()
df.shape # same as before: no Duplicate Rows
```

# 4. Cleaning Individual
Mapped book_table to 1/0 for easier processing

```python
def boolify(x):
    return 1 if x=="Yes" else 0
    
df["book_table"] = df["book_table"].apply(boolify)
df
```

#  5. Remove NaN

```python
df = df.dropna()
df.shape
```
